### Consulta de Endereço via ViaCEP

Este script Python demonstrará como consultar a API pública ViaCEP para obter informações de endereço a partir de um CEP e formatá-las de forma legível.

```python
import requests

print('Módulo requests importado com sucesso.')
```

Agora, vamos definir uma função que receberá um CEP, fará a requisição à API do ViaCEP e retornará o endereço formatado. Se o CEP for inválido ou não encontrado, a função informará o usuário.

```python
def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço.
    Retorna uma string formatada com o endereço ou uma mensagem de erro.
    """
    url = f"https://viacep.com.br/ws/{cep}/json/"
    try:
        response = requests.get(url)
        response.raise_for_status()  # Levanta um HTTPError para respostas de erro (4xx ou 5xx)
        data = response.json()

        if 'erro' in data and data['erro']:
            return f"Erro: CEP '{cep}' não encontrado ou inválido."
        else:
            endereco_formatado = (
                f"CEP: {data.get('cep')}\n"
                f"Logradouro: {data.get('logradouro')}\n"
                f"Complemento: {data.get('complemento')}\n"
                f"Bairro: {data.get('bairro')}\n"
                f"Localidade: {data.get('localidade')}\n"
                f"UF: {data.get('uf')}\n"
                f"IBGE: {data.get('ibge')}\n"
                f"GIA: {data.get('gia')}\n"
                f"DDD: {data.get('ddd')}\n"
                f"SIAFI: {data.get('siafi')}"
            )
            return endereco_formatado
    except requests.exceptions.HTTPError as e:
        return f"Erro HTTP: {e}"
    except requests.exceptions.ConnectionError as e:
        return f"Erro de conexão: Verifique sua internet. {e}"
    except requests.exceptions.Timeout as e:
        return f"Erro de timeout: A requisição demorou muito. {e}"
    except requests.exceptions.RequestException as e:
        return f"Erro inesperado na requisição: {e}"
    except ValueError:
        return "Erro: Resposta da API não é um JSON válido."
```

Por fim, um exemplo de como usar a função `consultar_cep`.

```python
# Exemplo de uso:
cep_valido = '01001000'
cep_invalido = '99999999'

print(f"\n--- Consultando CEP: {cep_valido} ---")
print(consultar_cep(cep_valido))

print(f"\n--- Consultando CEP: {cep_invalido} ---")
print(consultar_cep(cep_invalido))

# Você pode testar com outros CEPs:
# cep_outro = '13087000'
# print(f"\n--- Consultando CEP: {cep_outro} ---")
# print(consultar_cep(cep_outro))
```
